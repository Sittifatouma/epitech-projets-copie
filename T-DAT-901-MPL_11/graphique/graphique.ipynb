```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import csv
import seaborn as sns
```

```python
data = pd.read_csv(r'KaDo.csv')
```

```python
# Pour savoir quelle famille est la plus cher
def prixmoybyfam():
# Calculer le prix moyen par famille
    mean_price_by_family = data.groupby("FAMILLE")["PRIX_NET"].mean()

    # Créer le graphique à barres
    mean_price_by_family.plot(kind='bar')
    plt.xlabel("Famille")
    plt.ylabel("Prix moyen")
    plt.title("Prix moyen des produits par famille")
    plt.show()

prixmoybyfam()
```

```python
def prixmoybyfamwithecartype():
    # Calculer le prix moyen et l'écart type par famille
    mean_price_by_family = data.groupby("FAMILLE")["PRIX_NET"].mean()
    std_price_by_family = data.groupby("FAMILLE")["PRIX_NET"].std()

    # Créer le graphique à barres
    mean_price_by_family.plot(kind='bar', yerr=std_price_by_family)
    plt.xlabel("Famille")
    plt.ylabel("Prix")
    plt.title("Prix moyen des produits par famille avec écart-type")
    plt.show()
```

```python
prixmoybyfamwithecartype()
```

```python
# Calculer le nombre de produits par famille
product_count_by_family = data.groupby("FAMILLE")["LIBELLE"].size()

# Créer le graphique à barres
product_count_by_family.plot(kind='bar')
plt.xlabel("Famille")
plt.ylabel("Nombre de produits")
plt.title("Nombre de produits par famille")
plt.show()
```

```python
#Utiliser un graphique de ligne pour représenter l'évolution des ventes par mois, ce qui vous permettra de voir les tendances des ventes au fil du temps

def achatparmois():
    # Group by month and sum the sales
    sales_by_month = data.groupby("MOIS_VENTE")["PRIX_NET"].sum()

    # Plot the line chart
    sales_by_month.plot()
    plt.xlabel("Mois")
    plt.ylabel("Ventes (en euro)")
    plt.title("Évolution des ventes par mois")
    plt.show()
```

```python
achatparmois()
```

```python
def venteparfamille():
    # Group by product family and sum the sales
    sales_by_family = data.groupby("FAMILLE")["PRIX_NET"].sum()

    # Plot the bar chart
    sales_by_family.plot(kind='bar')
    plt.xlabel("Famille de produits")
    plt.ylabel("Ventes (en euro)")
    plt.title("Repartition des ventes par famille de produits")
    plt.show()
```

```python
venteparfamille()
```
