```python
PriceClient = df.loc[:, ["PRIX_NET", "CLI_ID"]]

MeanPrice = PriceClient.groupby("CLI_ID").mean()
MaxPrice = PriceClient.groupby("CLI_ID").max()
MinPrice = PriceClient.groupby("CLI_ID").min()
MeanPrice.head()
```
