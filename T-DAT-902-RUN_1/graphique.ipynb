```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_color = pd.read_csv("Couleur/fichier.csv")
df_date = pd.read_csv("Date/resultat4.csv")
df_title = pd.read_csv("Title/nouveau_fichier.csv")
df_visage = pd.read_csv("visage/nbvisage_emotion.csv")
df_train = pd.read_csv("../Multi_Label_dataset/train.csv")
df=pd.read_csv("data.csv")
```

Nombre de film par genre

```python
# Exclure les colonnes Id et Genre
data_genres = df_train.drop(['Id', 'Genre'], axis=1)

# Suppression des lignes vides
data_genres.dropna(inplace=True)

# Compter le nombre de films pour chaque genre
genre_counts = data_genres.sum()

# Crer un graphique  barres pour afficher les comptages
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Nombre de films')
plt.title('Nombre de films par genre')
plt.xticks(rotation=45)
plt.show()
```

Nombre de film par anne de sortie

```python
data_date= df

data_date['Data'] = data_date['Data']
counts = data_date['Data'].value_counts()

# Trier les valeurs par anne
counts = counts.sort_index()


# Crer un graphique  barres pour afficher les comptages
plt.figure(figsize=(10, 6))
counts.plot(kind='bar')
plt.xlabel('Anne')
plt.ylabel('Nombre de films')
plt.title('Nombre de films par anne')
plt.show()
```

Rpresentation de la couleur par genre de films

```python
# Crer une liste des genres disponibles
genres = ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy'
          , 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy'
          , 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'N/A', 'News', 'Reality-TV'
          , 'Romance', 'Sci-Fi', 'Short', 'Sport', 'Thriller', 'War', 'Western']  # Remplacez avec vos genres rels

# Crer une liste des couleurs disponibles
couleurs = ['noir', 'blanc', 'gris', 'rouge', 'vert', 'bleu', 'jaune', 'orange', 'rose', 'violet', 'marron'] # Remplacez avec vos couleurs relles

# Crer une figure et des sous-graphiques
fig, ax = plt.subplots()

# Pour chaque genre, obtenir les couleurs de films correspondantes
for i, genre in enumerate(genres):
    # Filtrer les donnes pour le genre actuel
    genre_df = df[df[genre] == 1]
    
    # Compter le nombre de films pour chaque couleur
    counts = [genre_df[genre_df['Couleur nominale'] == couleur].shape[0] for couleur in couleurs]
    
    # Tracer un graphique en barres pour le genre actuel
    ax.bar(couleurs, counts, label=genre)

# Configurer les axes et les lgendes
ax.set_xlabel('Couleurs')
ax.set_ylabel('Nombre de films')
ax.set_title('Couleurs de films par genre')
ax.legend()

# Afficher le graphe
plt.show()
```

```python

#Parcourir chaque genre de film
genres = df.columns[1:26]  # Les colonnes  partir de l'indice 1 contiennent les genres
for genre in genres:
    # Slectionner les donnes pour le genre spcifique
    genre_data = df[df[genre] == 1]

    # Compter les couleurs de film pour ce genre
    colors_count = genre_data['Couleur nominale'].value_counts()

    # Crer un graphique  barres pour les couleurs de film
    plt.figure()
    colors_count.plot(kind='bar')
    plt.title(f"Couleurs de film pour le genre {genre}")
    plt.xlabel("Couleur de film")
    plt.ylabel("Nombre de films")
    plt.show()

```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Parcourir chaque genre de film
genres = df.columns[1:26]  # Les colonnes  partir de l'indice 1 contiennent les genres
for genre in genres:
    # Slectionner les donnes pour le genre spcifique
    genre_data = df[df[genre] == 1]

    # Compter les motions exprimes pour ce genre
    emotions_count = genre_data['emotions'].value_counts()

    # Crer un graphique  barres pour les motions exprimes
    plt.figure()
    emotions_count.plot(kind='bar')
    plt.title(f"motions exprimes pour le genre {genre}")
    plt.xlabel("motion")
    plt.ylabel("Nombre de films")
    plt.show()


```
