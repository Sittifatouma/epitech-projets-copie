```python
import cv2
import numpy as np
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
def get_couleur_dominante(image_path):
    # Charger l'image avec OpenCV
    image = cv2.imread(image_path)

    # Convertir l'image de BGR à RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Redimensionner l'image pour accélérer le calcul de l'histogramme
    resized_image = cv2.resize(image_rgb, (100, 100), interpolation=cv2.INTER_AREA)

    # Aplatir l'image en un tableau 2D
    flattened_image = resized_image.reshape((-1, 3))

    # Calculer l'histogramme des couleurs
    histogram = np.zeros((256, 256, 256), dtype=np.uint8)
    for pixel in flattened_image:
        histogram[pixel[0], pixel[1], pixel[2]] += 1

    # Trouver l'indice du pixel le plus fréquent dans l'histogramme
    indice_max = np.argmax(histogram)
    couleur_dominante = np.unravel_index(indice_max, histogram.shape)

    return couleur_dominante
```

```python
def convertir_en_couleur_nominale(couleur_rgb):
    # Définir une liste de couleurs nominatives
    couleurs_nominatives = ['noir', 'blanc', 'gris', 'rouge', 'vert', 'bleu', 'jaune', 'orange', 'rose', 'violet', 'marron']

    # Définir les valeurs RVB correspondantes aux couleurs nominatives
    couleurs_rgb = np.array([
        [0, 0, 0],     # noir
        [255, 255, 255],   # blanc
        [128, 128, 128],   # gris
        [255, 0, 0],   # rouge
        [0, 255, 0],   # vert
        [0, 0, 255],   # bleu
        [255, 255, 0],   # jaune
        [255, 165, 0],   # orange
        [255, 192, 203],   # rose
        [128, 0, 128],   # violet
        [165, 42, 42]   # marron
    ])

    # Créer un modèle Nearest Neighbors et ajuster les données RVB
    knn = NearestNeighbors(n_neighbors=1)
    knn.fit(couleurs_rgb)

    # Trouver le voisin le plus proche des valeurs RVB données
    _, indice_plus_proche = knn.kneighbors([couleur_rgb])
    indice_plus_proche = indice_plus_proche[0][0]

    # Récupérer le nom de couleur correspondant à l'indice
    couleur_nominale = couleurs_nominatives[indice_plus_proche]

    return couleur_nominale
```

```python
import cv2
import numpy as np
from sklearn.neighbors import NearestNeighbors
import os
import csv

def get_couleur_dominante(image_path):
    # Charger l'image avec OpenCV
    image = cv2.imread(image_path)

    # Convertir l'image de BGR à RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Redimensionner l'image pour accélérer le calcul de l'histogramme
    resized_image = cv2.resize(image_rgb, (100, 100), interpolation=cv2.INTER_AREA)

    # Aplatir l'image en un tableau 2D
    flattened_image = resized_image.reshape((-1, 3))

    # Calculer l'histogramme des couleurs
    histogram = np.zeros((256, 256, 256), dtype=np.uint8)
    for pixel in flattened_image:
        histogram[pixel[0], pixel[1], pixel[2]] += 1

    # Trouver l'indice du pixel le plus fréquent dans l'histogramme
    indice_max = np.argmax(histogram)
    couleur_dominante = np.unravel_index(indice_max, histogram.shape)

    return couleur_dominante


def convertir_en_couleur_nominale(couleur_rgb):
    # Définir une liste de couleurs nominatives
    couleurs_nominatives = ['noir', 'blanc', 'gris', 'rouge', 'vert', 'bleu', 'jaune', 'orange', 'rose', 'violet', 'marron']

    # Définir les valeurs RVB correspondantes aux couleurs nominatives
    couleurs_rgb = np.array([
        [0, 0, 0],     # noir
        [255, 255, 255],   # blanc
        [128, 128, 128],   # gris
        [255, 0, 0],   # rouge
        [0, 255, 0],   # vert
        [0, 0, 255],   # bleu
        [255, 255, 0],   # jaune
        [255, 165, 0],   # orange
        [255, 192, 203],   # rose
        [128, 0, 128],   # violet
        [165, 42, 42]   # marron
    ])

    # Créer un modèle Nearest Neighbors et ajuster les données RVB
    knn = NearestNeighbors(n_neighbors=1)
    knn.fit(couleurs_rgb)

    # Trouver le voisin le plus proche des valeurs RVB données
    _, indice_plus_proche = knn.kneighbors([couleur_rgb])
    indice_plus_proche = indice_plus_proche[0][0]

    # Récupérer le nom de couleur correspondant à l'indice
    couleur_nominale = couleurs_nominatives[indice_plus_proche]

    return couleur_nominale

# Chemin vers le dossier contenant les images
dossier_images = r"/Users/dimitriraymond/Documents/Epitech/Semestre 2/DAT/data/Movie_Poster_Dataset"

# Liste pour stocker les résultats
resultats = []

# Parcourir les fichiers et dossiers dans le dossier d'images
for dossier, sous_dossiers, fichiers in os.walk(dossier_images):
    for fichier in fichiers:
        # Vérifier si le fichier est une image JPG
        if fichier.endswith(".jpg") or fichier.endswith(".jpeg"):
            # Chemin complet vers l'image
            chemin_image = os.path.join(dossier, fichier)
            
            # Obtenir la couleur dominante
            couleur_dominante = get_couleur_dominante(chemin_image)
            
            # Convertir en couleur nominale
            couleur_nominale = convertir_en_couleur_nominale(couleur_dominante)
            
            # Ajouter le titre de l'image et la couleur nominale à la liste des résultats
            titre_image = fichier  # Utiliser le nom du fichier comme titre
            resultats.append([titre_image, couleur_nominale])

# Écrire les résultats dans un fichier CSV
chemin_fichier_csv = r"/Users/dimitriraymond/Documents/Epitech/Semestre 2/DAT/resultc/fichier.csv"
with open(chemin_fichier_csv, mode='w', newline='') as fichier_csv:
    writer = csv.writer(fichier_csv)
    writer.writerow(["Titre de l'image", "Couleur nominale"])  # Écrire l'en-tête du fichier CSV
    writer.writerows(resultats)  # Écrire les résultats

print("Terminé. Les résultats ont été enregistrés dans", chemin_fichier_csv)
```
