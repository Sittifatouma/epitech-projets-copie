```python
def get_predicted_genres(prediction):
    genres = ['Comedy', 'Documentary', 'Horror', 'N/A', 'Short', 'Sport', 'Western', 'Action_Adventure_Sci-Fi',
              'Animation_Family_Fantasy', 'Music_Musical', 'Crime_Mystery_Thriller', 'Drama_Romance',
              'History_Biography_War', 'Reality_TV_News']
    
    predicted_genres = []
    for i, genre in enumerate(genres):
        if prediction[0][i] == 1:
            predicted_genres.append(genre)
    
    return predicted_genres
```
